### hgt.mat

```python
import scipy.io as sio

matfile = 'hgt.mat'
data = sio.loadmat(matfile)
```

用dict的形式存储mat数据

```python
print(data)
```

读取'hgt_yy'键所对应的值

```python
hgt_list = data['hgt_yy']
print("hgt_yy的数据类型为", hgt_list.shape)
print("所有数据如下：\n%s" %hgt_list)
```

第一行数据显示


```python
print("第一行数据类型为： %s" %hgt_list[0].shape)
print("第一行数值为： \n%s" %hgt_list[0])
```

保存为.csv文件

#### 将hgt.mat里面的数据存储到csv

调用`csv`库进行写入，`writerow`是行读入，即读入一行，`writerows`是多行读入
- *line5*:在首行写入标题，`year`和`key{}`,`.format(i)`是对`key{}`里面赋值
- *line7*:迭代逐行写入，`hgt_list[j]`表示第j行，`list()`是对其切片写入

```python
import csv
year_start = 1951
with open("hgt.csv", "w") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['year']+['key{}'.format(i) for i in range(208)])
    for j in range(66):
        writer.writerow([year_start+j]+ list(hgt_list[j]))
```
