### 2#炉数据清洗

1.从Excel文件中读取每个工作表中影响特征，然后拼接到一个工作表中。
跟1#炉不同的是里面的时间不是整数的，所以和二噁英的结果不吻合，无法直接按照时间取拼接，所以需要先将X拼接起来，按照时间去合并成整点的。

```python
from functools import reduce

import pandas as pd

ereying='二噁英工控数据（2#炉）.xlsx'#二噁英数据

data1=pd.read_excel(ereying,sheetname=u'尾部烟道进口烟气温度')
data2=pd.read_excel(ereying,sheetname=u'尾部烟道出口烟气温度')
data3=pd.read_excel(ereying,sheetname=u'炉膛出口O2浓度1')
data4=pd.read_excel(ereying,sheetname=u'炉膛出口O2浓度2')
data5=pd.read_excel(ereying,sheetname=u'CO的浓度')
data6=pd.read_excel(ereying,sheetname=u'炉膛第一烟道出口烟气温度')
data7=pd.read_excel(ereying,sheetname=u'炉膛第一烟道入口烟气温度')
data8=pd.read_excel(ereying,sheetname=u'活性炭喷射量')
data9=pd.read_excel(ereying,sheetname=u'一次风机实际运行电流')
data10=pd.read_excel(ereying,sheetname=u'二次风机实际运行电流')
data11=pd.read_excel(ereying,sheetname=u'锅炉蒸汽量')


print("锅炉蒸汽量", data11)

dfs=[data1,data2,data3,data4,data5,data6,data7,data8,data9,data10,data11]
df_final=reduce(lambda left,right:pd.merge(left,right,on='时间',how='outer'), dfs)
names = [u'时间',u'尾部烟道进口烟气温度',u'炉尾部烟道出口烟气温度',u'炉膛出口O2浓度1',u'炉膛出口O2浓度2',u'CO的浓度',u'炉膛第一烟道出口烟气温度',
         u'炉膛第一烟道入口烟气温度',u'活性炭喷射量',u'一次风机实际运行电流',u'二次风机实际运行电流',u'锅炉蒸汽量']
df_final.columns = names
df_final.to_excel('X.xlsx')
```

2.`X.xlsx`是所有特征拼接到一个sheet中，下面需要做的是按照时间将数据相加每小时保留一个值

```python
import pandas as pd
ereying=pd.DataFrame(pd.read_excel('X.xlsx'))
ereying=ereying.set_index(u'时间')

ereying2=ereying.resample('H').sum()

ereying2.to_excel('X_H.xlsx',encoding='utf-8')
```

3.然后按照1的拼接方法将数据和标签拼接到一张表上面

```python
from functools import reduce
import pandas as pd

ereying_x = 'X_H.xlsx'
data1 = pd.read_excel(ereying_x)
ereying_y = '二噁英工控数据（2#炉）.xlsx'
data2 = pd.read_excel(ereying_y, sheetname=u'二噁英排放浓度')
print(data1)
print(data2)

dfs = [data1, data2]
df_final = reduce(lambda left,right:pd.merge(
                            left, right, on=u'时间',how='outer'), dfs)
names = [u'时间',u'尾部烟道进口烟气温度',u'炉尾部烟道出口烟气温度',u'炉膛出口O2浓度1',u'炉膛出口O2浓度2',u'CO的浓度',u'炉膛第一烟道出口烟气温度',
         u'炉膛第一烟道入口烟气温度',u'活性炭喷射量',u'一次风机实际运行电流',u'二次风机实际运行电流',u'锅炉蒸汽量',u'二噁英排放浓度']
df_final.columns = names
df_final.to_excel('X_y.xlsx')
```

加载已有模型对2#炉数据进行拟合

```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-
# Created by XiaoYu on 18-3-28
import pandas as pd
# import numpy as np
from sklearn import metrics
import pickle
from xgboost.sklearn import XGBRegressor
# from sklearn.preprocessing import StandardScaler
# from clean_data import prep_water_data,normalize_water_data,normalize_data,delete_null_date
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV, cross_val_score
from sklearn.model_selection import TimeSeriesSplit

def GDBTest(X, y, clf):
    test_preds = pd.DataFrame({"label": y})
    test_preds['y_pred'] = clf.predict(X)
    print("验证集的个数", len(test_preds))
    stdm = metrics.r2_score(test_preds['label'], test_preds['y_pred'])
    loss = -cross_val_score(clf, X, y, cv=10, scoring='neg_mean_squared_error')
    print("交叉验证平方差", loss.mean())
    cross_val = cross_val_score(clf, X, y, cv=10)
    print("交叉验证的结果： ",cross_val)    
    return stdm, test_preds

if __name__ == "__main__":

    #验证
    #读取验证集数据
    X_data_test = pd.read_csv('X_data.csv')[:]
    Y_data_test = pd.read_csv('Y_label.csv')[:]
    Y_data_test = Y_data_test[u'二噁英排放浓度']

    print("输入数据类型：", X_data_test.shape)
    print("标签类型：", Y_data_test.shape)
    X = X_data_test.as_matrix()
    y = Y_data_test.as_matrix()
    out = 'frazee'
    path = './model/' + out + '_xgb.pkl'
    #模型载入并测试
    with open(path, 'rb') as f:
        clf = pickle.load(f)
        stdm, prediction = GDBTest(X, y, clf)
    print("验证r2_score的值：", stdm)
    print("输出验证结果：", prediction)


    #3.可视化
# 绘制实际结果和预测结果
    import matplotlib.pyplot as plt
    import numpy as np


    x = np.arange(len(prediction))
    plt.scatter(x[0:10], prediction['y_pred'][0:10])
    plt.scatter(x[0:10], prediction['label'][0:10], c='r')
    plt.xlabel('number')
    plt.ylabel('value')
#     plt.ylim(0, 0.02)
    plt.grid(True)
    plt.legend(labels=['pre', 'label'], loc='best')
    plt.show()
```
